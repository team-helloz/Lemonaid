```python
import pandas as pd
```

```python
d_df = pd.read_csv('./data/disease.csv', encoding='cp949')
d_df.index = d_df.index + 1
```

```python
d_df = d_df.fillna('')
d_df = d_df[['disease_no', 'disease_name', 'disease_definition', 'disease_cause', 'disease_symptom_description', 'disease_diagnosis', 'disease_treatment', 'disease_caution']]
d_df
```

```python
d_df['disease_script'] = d_df['disease_definition']+" "+d_df['disease_cause']+" "+d_df['disease_symptom_description']+" "+d_df['disease_diagnosis']+" "+d_df['disease_treatment']+" "+d_df['disease_caution']
```

```python
d_df = d_df[['disease_no', 'disease_name', 'disease_script']]
d_df
```

```python
temp_df = pd.DataFrame(d_df['disease_name'].str.split("(", 1).tolist())
temp_df = temp_df[[0]]
d_df['disease_name'] = temp_df.copy()
d_df
```

```python
d_df.to_csv('./output/disease_script.csv', index=False, encoding="utf-8-sig")
```

```python
from konlpy.tag import Komoran
komoran=Komoran()
```

```python
nouns = []
for i in range(1, 1273 + 1):
    if d_df['disease_script'][i] != '':
        try:
            noun = komoran.nouns(d_df['disease_script'][i])
        except:
            noun =[]
    t_df = pd.DataFrame(noun)
    t_df.to_csv('./output/disease_nouns_'+str(i)+'.csv', index=False, encoding="utf-8-sig", header= None)
# d_df
```

```python
d_df['disease_nouns'] = nouns
d_df
```

```python
d_df.to_csv('./output/disease_nouns.csv', index=False, encoding="utf-8-sig")
```
