```python
import pandas as pd
from ast import literal_eval
```

```python
dis_df = pd.read_csv('./disease.csv',d='UTF-8')
dis_df = dis_df.fillna('')

dis_df.index = dis_df.index + 1
dis_df
```

```python
disease_symptom = []

for i in range(1, len(dis_df)+ 1):
    try:
        syms = literal_eval(dis_df.loc[i]['disease_symptoms'])
    except:
        syms = []
    for sym in syms:
        disease_symptom.append([i, dis_df.loc[i]['disease_name'], sym])
```

```python
dis_syms_df = pd.DataFrame(disease_symptom)
dis_syms_df.to_csv('./symptom_disease.csv', index=False, encoding="utf-8-sig")
```

```python
disease_related = []

for i in range(1, len(dis_df)+ 1):
    try:
        rels = literal_eval(dis_df.loc[i]['related_disease_name'])
    except:
        rels = []
    for rel in rels:
        disease_related.append([i, dis_df.loc[i]['disease_name'], rel])
```

```python
dis_rel_df = pd.DataFrame(disease_related)
dis_rel_df.to_csv('./related_disease.csv', index=False, encoding="utf-8-sig")
```

```python
disease_subjects = []

for i in range(1, len(dis_df)+ 1):
    try:
        subs = literal_eval(dis_df.loc[i]['disease_subjects'])
    except:
        subs = []
    for sub in subs:
        disease_subjects.append([i, dis_df.loc[i]['disease_name'], sub])
```

```python
dis_rel_df = pd.DataFrame(disease_subjects)
dis_rel_df.to_csv('./disease_subjects.csv', index=False, encoding="utf-8-sig")
```
