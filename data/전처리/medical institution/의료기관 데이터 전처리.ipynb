```python
import pandas as pd
import numpy as np
```

```python
# 병원 데이터 불러오기
# 데이터 인코딩 방식 옵션으로 설정
data = pd.read_csv("./data/hospitol.csv", encoding='cp949')
data
```

```python
# 필요없는 열 삭제
data = data.drop(data.columns[13:27], axis=1)
data
```

```python
# 진료시간 정보를 담고 있는 csv 불러오기
hospitol_time = pd.read_csv("./data/timetable.csv", encoding='cp949')
hospitol_time
```

```python
# 요양기관명을 기준으로 병합
# data <= hospitol_time
hospitol = pd.merge(data, hospitol_time, left_on="암호화요양기호", right_on="암호화요양기호", how='left')
```

```python
cnt = 0
for column in hospitol:
    print(column, cnt)
    cnt += 1
```

```python
hospitol = hospitol.drop(hospitol.columns[[8, 9, 15, 16, 17, 18, 30, 31, 32, 33]], axis=1)
```

```python
cnt = 0
for column in hospitol:
    print(column, cnt)
    cnt += 1
```

```python
# 좌표없는 값 api 활용 데이터 값 채우기
import requests
import json
    

url = 'http://api.vworld.kr/req/address?'
params = 'service=address&request=getcoord&version=2.0&crs=epsg:4326&refine=true&simple=false&format=json&type='
road_type = 'ROAD'
address = '&address='
keys = '&key='
primary_key = 'B333E8A4-9331-39D8-B874-07E272497C22'


for adr, idx in zip(hospitol[hospitol['x좌표'].isna()]['주소'], hospitol[hospitol['x좌표'].isna()].index):
    page = requests.get(url + params + road_type + address + adr + keys + primary_key)
    json_data = page.json()
    status = json_data['response']['status']
    if status == 'NOT_FOUND' or status == 'ERROR':
        print(adr, idx)
        continue
    x = json_data['response']['result']['point']['x']
    y = json_data['response']['result']['point']['y']
    hospitol.loc[hospitol.index == idx, ('x좌표', 'y좌표')] = (x, y)
```
