```python
shape_dict = dict()
info_dict = dict()
```

```python
# for i in medicine_shape.iloc[:, [1]]:
for i in range(4390):
    if medicine_info["제품명"][i] in info_dict.keys():
        info_dict[medicine_info["제품명"][i]] += 1
    else:
        info_dict[medicine_info["제품명"][i]] = 1

for i in range(24316):
    if medicine_shape["품목명"][i] in shape_dict.keys():
        shape_dict[medicine_shape["품목명"][i]] += 1
    else:
        shape_dict[medicine_shape["품목명"][i]] = 1
```
