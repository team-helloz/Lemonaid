```python
import os, shutil
import time
import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
```

```python
data = pd.read_csv("./data/medicine_shape.csv")
```

```python
data.head()
```

```python
md_img = []

for num, url in zip(data["품목일련번호"], data["큰제품이미지"]):
    md_img.append([num, url])
print(len(md_img))
```

```python
filepath = 'C:\\Users\\multicampus\\Desktop\\new'

options = webdriver.ChromeOptions()

options.add_argument('headless')

options.add_experimental_option("prefs", {
  "download.default_directory": "C:\\Users\\multicampus\\Desktop\\new",
  "download.prompt_for_download": False,
  "download.directory_upgrade": True,
  "safebrowsing.enabled": True
})

start = time.time()

driver = webdriver.Chrome(options=options)
for i in range(1, 1000):
    if i % 100 == 0:
        print(i)
    key = md_img[i][0]
    val = md_img[i][1]
    
    driver.get(val)
    time.sleep(2)
    
    newfilename = './img/{}.jpg'.format(key)

    filename = max([filepath + '\\' + f for f in os.listdir(filepath)], key=os.path.getctime)
    shutil.move(os.path.join(filepath, filename), newfilename)

driver.close()
print("time :", time.time() - start)
```
