```python
import pandas as pd
from ast import literal_eval
```

```python
dis_df = pd.read_csv('./disease.csv', encoding='cp949')
dis_df = dis_df.fillna('')
dis_df.index = dis_df.index + 1
print(dis_df)
dis_df
```

```python
sym_df = pd.read_csv('./symptom.csv', encoding='cp949')
sym_df
```

```python
dis_sym_df = pd.read_csv('./disease_symptom.csv', encoding='cp949')
dis_sym_df.index = dis_sym_df.index + 1
dis_sym_df
```

```python
dis_sym_df = pd.merge(dis_sym_df, sym_df, left_on='symptom', right_on='symptom_name').drop(columns='symptom')
dis_sym_df.index = dis_sym_df.index + 1

dis_sym_df.to_csv('./disease_symptom.csv', index=True, encoding='utf-8-sig')
```

```python
dis_rel_df = pd.read_csv('./disease_related.csv', encoding='cp949')
dis_rel_df.index = dis_rel_df.index + 1
dis_rel_df
```

```python
dis_df['temp_name'] = dis_df.disease_name.str.split('(').str[0]
temp_dis_df = dis_df[['disease_no', 'disease_name', 'temp_name']]
temp_dis_df


dis_rel_df = pd.merge(dis_rel_df, temp_dis_df, left_on='related_disease_name', right_on='temp_name')
dis_rel_df.index = dis_rel_df.index + 1
# dis_rel_df[['disease_no_x', 'disease_name_x', 'disease_no_y', 'related_disease_name']]
```

```python
dis_rel_df = dis_rel_df.rename(columns={'disease_no_x':'disease_no', 'disease_name_x': 'disease_name', 'disease_no_y': 'related_disease_no'})
dis_rel_df = dis_rel_df[['disease_no', 'disease_name', 'related_disease_no', 'related_disease_name']]
dis_rel_df

dis_rel_df.to_csv('./related_disease_완.csv', index=True, encoding='utf-8-sig')
```
