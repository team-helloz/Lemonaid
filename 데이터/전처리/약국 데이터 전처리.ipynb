```python
import pandas as pd
```

```python
# 약국 데이터
# 영업시간 데이터
pmc = pd.read_csv("./data/pharmacy.csv", encoding='cp949')
tt = pd.read_csv("./data/timetable.csv", encoding='cp949')
```

```python
pmc.head()
```

```python
# 요양기관명을 기준으로 병합
# pmc <= tt
data = pd.merge(pmc, tt, left_on="암호화요양기호", right_on="암호화요양기호", how='left')
```

```python
# 필요없는 컬럼 삭제
data = data.drop(data.columns[[8, 9, 12, 15, 16, 17, 18, 30, 31, 32, 33]], axis=1)
```

```python
cnt = 0
for column in data:
    print(column, cnt)
    cnt += 1
```

```python
# 좌표가 없는 데이터 행의 주소 출력
data[data['x좌표'].isna()]['주소']
```

```python
# 좌표없는 값 api 활용 데이터 값 채우기
import requests
import json
    

url = 'http://api.vworld.kr/req/address?'
params = 'service=address&request=getcoord&version=2.0&crs=epsg:4326&refine=true&simple=false&format=json&type='
road_type = 'ROAD'
address = '&address='
keys = '&key='
primary_key = 'B333E8A4-9331-39D8-B874-07E272497C22'


for adr, idx in zip(data[data['x좌표'].isna()]['주소'], data[data['x좌표'].isna()].index):
    page = requests.get(url + params + road_type + address + adr + keys + primary_key)
    json_data = page.json()
    status = json_data['response']['status']
    if status == 'NOT_FOUND' or status == 'ERROR':
        continue
    x = json_data['response']['result']['point']['x']
    y = json_data['response']['result']['point']['y']
    data.loc[data.index == idx, ('x좌표', 'y좌표')] = (x, y)
    print(x, y)
    print(data.loc[idx]['x좌표'], data.loc[idx]['y좌표'])
    print('----------------')


```
