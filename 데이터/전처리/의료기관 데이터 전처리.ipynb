```python
import pandas as pd
```

```python
# 병원 데이터 불러오기
# 데이터 인코딩 방식 옵션으로 설정
data = pd.read_csv("./data/hospitol.csv", encoding='cp949')
data
```

```python
# 필요없는 열 삭제
data = data.drop(data.columns[13:27], axis=1)
data
```

```python
# 진료시간 정보를 담고 있는 csv 불러오기
hospitol_time = pd.read_csv("./data/timetable.csv", encoding='cp949')
hospitol_time
```

```python
# 요양기관명을 기준으로 병합
# data <= hospitol_time
pd.merge(data, hospitol_time, left_on="암호화요양기호", right_on="암호화요양기호", how='left')
```
